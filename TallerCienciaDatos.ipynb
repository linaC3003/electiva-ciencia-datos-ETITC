```python
import pandas as pd
url = "https://raw.githubusercontent.com/databricks/Spark-The-Definitive-Guide/refs/heads/master/data/retail-data/all/online-retail-dataset.csv?classId=cd0b6b9e-28f2-424f-9517-30a93a11d74a"
data = pd.read_csv(url, encoding='latin1')
# Mostrar las primeras cinco filas
data.head()
```

```python
# Verificar valores faltantes
print(data.isnull().sum())

# Obtener estadísticas resumidas
print(data.describe())

# Filtrar datos para mostrar solo ventas exitosas (sin valores negativos en la cantidad)
data = data[data['Quantity'] > 0]

# Crear un gráfico para visualizar las ventas por país
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(y='Country', data=data, order=data['Country'].value_counts().index)
plt.title('Número de Transacciones por País')
plt.xlabel('Número de Transacciones')
plt.ylabel('País')
plt.show()
```

En la gráfica proporcionada, se visualiza el número de transacciones realizadas por país en un conjunto de datos filtrado para considerar únicamente las ventas exitosas, es decir, aquellas con cantidades positivas (sin devoluciones o valores negativos).

Se representa el número de transacciones realizadas, mostrado como barras horizontales para cada país.

La mayor cantidad de transacciones se concentra en el Reino Unido (United Kingdom), lo cual es evidente por la longitud considerablemente mayor de su barra en comparación con los demás países.
Otros países como Alemania (Germany), Francia (France), y Irlanda (EIRE) presentan un menor número de transacciones, aunque significativamente más que otros países en la lista.
Algunos países aparecen con muy pocas transacciones, representadas por barras apenas visibles.

El gráfico indica que el Reino Unido es el principal mercado en este conjunto de datos.

```python
import matplotlib.pyplot as plt

# Histograma para la cantidad de ventas
data['Quantity'].hist()
plt.xlabel('Cantidad de Productos Vendidos')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Cantidad de Productos Vendidos')
plt.show()
```

En esta segunda gráfica se presenta un histograma que ilustra la distribución de la cantidad de productos vendidos en el conjunto de datos.

Se representa la cantidad de productos vendidos en cada transacción. La escala abarca desde 0 hasta aproximadamente 80,000 y se muestra la frecuencia o el número de transacciones que tienen una determinada cantidad de productos vendidos.

Interpretación:
La mayoría de las ventas corresponden a pequeñas cantidades de productos.
Los valores extremos (grandes cantidades de productos vendidos en una sola transacción) son raros y poco representativos de la mayoría del conjunto de datos.

```python
import plotly.express as px

fig = px.scatter(data, x="Quantity", y="UnitPrice", color="Country", title="Relación entre Cantidad y Precio Unitario por País")
fig.show()
```

Esta gráfica de dispersión muestra la relación entre la cantidad de productos vendidos (Quantity) y el precio unitario (UnitPrice) para transacciones realizadas en diferentes países.

(Quantity): Representa la cantidad de productos vendidos en una transacción. Los valores oscilan entre 0 y más de 10 K.

(UnitPrice): Representa el precio unitario del producto en la transacción. Los valores van desde negativos (probablemente errores o devoluciones) hasta precios elevados.

La mayoría de los puntos están concentrados cerca del origen, con valores bajos tanto en cantidad como en precio unitario.
Es difícil identificar un patrón claro debido a la escala amplia, no todos los colores de cada pais se logra distinguir correctamente. Se observa mas participacion de United Kingdom vendiendo mayor cantidad de productos y a un mayor precio.


```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

p = figure(title="Cantidad de Productos Vendidos por Fecha", x_axis_label='Fecha', y_axis_label='Cantidad')
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
sales_per_date = data.groupby(data['InvoiceDate'].dt.date)['Quantity'].sum()

p.line(x=sales_per_date.index, y=sales_per_date.values, legend_label="Ventas Diarias", line_width=2)
show(p)
```

Este ultimo grafico representa la "Cantidad de Productos Vendidos por Fecha".

El código carga una columna llamada InvoiceDate de un conjunto de datos (probablemente un DataFrame de Pandas) y la convierte a un formato de fecha.
Luego agrupa los datos por fecha y suma la cantidad de productos vendidos (Quantity) para cada día.

Eje X (Fecha): Muestra los valores de la columna InvoiceDate, que están en un formato numérico de tiempo.
Eje Y (Cantidad): Representa la cantidad total de productos vendidos en cada fecha específica.
La línea azul muestra la cantidad diaria de ventas, etiquetada como "Ventas Diarias".

Se observa un pico significativo en las ventas en ciertos puntos, lo que podría indicar fechas de alta demanda o eventos especiales.
En general, hay fluctuaciones diarias en la cantidad de ventas, con variaciones grandes en algunos días.
Este gráfico ayuda a visualizar cómo varían las ventas diarias en el tiempo, permitiendo identificar patrones, picos de alta demanda y posibles estacionalidades.
